Исследование паралелльной версии алгоритма

Исследование проводится на входных данных содержащих:

Число процессоров: 100, число работ: 100000, минимальная длительность одной работы: 1, максимальная длительность одной работы: 100    

```python
data = """1 22.7269 53891
2 41.0802 52774
3 32.3875 53444
4 21.7353 53592
5 24.158 52763
6 34.5513 53203"""
```

```python
data_split = [i.split() for i in data.split('\n')]
mulproc_data = pd.DataFrame(data_split, columns=["num_proc", "time", "target_func"], dtype=float)
mulproc_data
```

```python
fig, ax = plt.subplots()
fig.set_figheight(5)
fig.set_figwidth(10)
ax.plot(mulproc_data["num_proc"], mulproc_data["time"], marker='o')
ax.set_xlabel("Число нитей")
ax.set_ylabel("Время вычисления (с)")
ax.grid()

ax2=ax.twinx()
ax2.plot(mulproc_data["num_proc"], mulproc_data["target_func"], color="green", marker='o')
ax2.set_ylabel("Значение ЦФ")

plt.title("Зависимость времени работы алгоритма от числа нитей")

plt.show()
```
